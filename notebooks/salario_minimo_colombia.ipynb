# Salario mínimo en Colombia (últimos 8 años)

## Datos del salario minimo en colombia en los ultimos 8 años
A continuacion se muestran los datos del salario minimo en los ultimos 8 años en miles COP

```python
# Salarios minimos en colombia en COP
salarios = [737717, 781242, 828116, 877803, 908526, 1000000, 1160000, 1300000]
```

## Calcular la media del salario

```python
# Se calcula la media de los salarios
media = sum(salarios) / len(salarios)

print("Media del salario:", media)
```

El resultado de la media nos muestra que el salario promedio tuvo una tendencia creciente

## Determinar la media

```python
# Ordenamos los datos
valores = sorted(salarios)

# Obtenemos la cantidad de datos
n = len(valores)

# Como es par, se promedian los dos del centro
mediana = (valores[n//2 - 1] + valores[n//2]) / 2

print("Mediana:", mediana)
```

La mediana confirma que los valores están equilibrados sin grandes distorsiones

## ¿Cuál es el incremento máximo entre dos años consecutivos?

```python
# Se inicializa la variable para el incremento maximo
maximo = 0

# Recorrer comparando año actual con el siguiente
for i in range(len(salarios) - 1):

    # Calcular el incremento entre dos años consecutivos
    incremento = salarios[i+1] - salarios[i]
    
    # Verificar si este incremento es mayor que el actual maximo
    if incremento > maximo:
        maximo = incremento

print("Incremento maximo:", maximo)
```

El incremento maximo refleja un aumento significativo en uno de los años recientes

## ¿Cuántos años superan el promedio?

```python
# Se inicializa el contador
contador = 0

# Se busca en la lista de valores de los salarios
for salario in salarios:

    # Si el salario es mayor que la media se añade un numero
    if salario > media:
        contador += 1

print("Años que superan el promedio:", contador)
```

Varios años superan el promedio, lo que indica que el salario ha venido aumentando progresivamente
