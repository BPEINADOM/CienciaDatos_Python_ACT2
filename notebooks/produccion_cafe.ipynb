# Producción de café (millones de sacos)

## Datos Historicos de produccion nacional

```python
# Produccion de cafe en millones de sacos
cafe = [14.2, 13.8, 14.5, 13.3, 12.6, 13.9, 11.5, 12.8, 14.0, 13.4]
```

## Calcular media y mediana de produccion

```python
# Se calcula la media de la produccion de cafe
media = sum(cafe) / len(cafe)

print("Media de la produccion de cafe:", media)
```

```python
# Ordenamos los datos
valores = sorted(cafe)

# Obtenemos la cantidad de datos
n = len(valores)

# Al ser par, se promedian los dos valores centrales
mediana = (valores[n//2-1] + valores[n//2]) / 2

print("Mediana de la produccion de cafe:", mediana)
```

La produccion promedio de cafe se mantiene alrededor de valores estables, confirmandose que no hay cambios extremos en la mediana

## Determinar el año con mayor produccion

```python
# Se inicializa el valor 
maximo = cafe[0]

# Se buscan los valores en la lista de datos para encontrar el valor maximo
for valor in cafe:
    if valor > maximo:
        maximo = valor

print ("Mayor produccion de cafe:", maximo)
```

Gracias al resultado de la operacion se puede observar el mejor año de produccion de cafe con un numero importante

## Determinar el año con menor produccion

```python
# Se inicializa el valor 
minimo = cafe[0]

# Se buscan los valores en la lista de datos para encontrar el valor minimo
for valor in cafe:
    if valor < minimo:
        minimo = valor

print ("Menor produccion de cafe:", minimo)
```

El minimo indica una caida importante de la produccion de cafe

## ¿Cuántos años estuvieron por encima de 13 millones?

```python
# Contador de años con produccion mayor a 13
contador = 0

# Recorrer la lista
for valor in cafe:
    if valor > 13:
        contador += 1

print("Años por encima de 13 millones:", contador)
```

La mayoria de los años estan por encima de 13 millones, lo que refleja una produccion relativamente constante.
