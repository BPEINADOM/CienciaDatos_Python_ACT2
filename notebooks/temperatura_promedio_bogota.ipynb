# Temperaturas promedio en Bogotá (°C)

## Datos de temperaturas mensuales en Bogota

```python
# Datos de temperatura mensuales en Bogota (°C)
temperaturas = [14, 15, 14, 13, 13, 12, 12, 13, 14, 14, 15, 14]
```

## Calcular la media y mediana

```python
# Se calcula la media de las temperaturas
media = sum(temperaturas) / len(temperaturas)

# Se imprime el resultado de la media
print("Media de temperaturas:", media)
```

```python
# Se ordenan los valores de las temperaturas
valores = sorted(temperaturas)

# Obtenemos la cantidad de datos que hay
n = len(valores)

# Se calcula la mediana usando la formula par, al haber 12 valores, se buscan los valores en la mitad y se promedian
mediana = (valores[n//2 - 1] + valores[n//2]) / 2

print("Mediana de temperaturas:", mediana)
```

La media y la mediana muestran que la temperatura promedio en bogota es alrededor de 14°c

## Determinar la moda

```python
# Se inicializan las variables de moda y maxima repetiticion del valor
moda = temperaturas[0]
max = 0

# Se busca el valor que mas se repite
for valor in temperaturas:

    # Cuenta cuantas veces aparece ese valor
    repeticiones = temperaturas.count(valor)

    # Si se repite mas que el valor anterior se actualiza la moda
    if repeticiones > max:
        max = repeticiones
        moda = valor

print("Moda de las temperturas:", moda)
```

La moda indica que 14°C es la temperatura mas frecuente

## ¿Cuál fue la temperatura máxima y mínima?

```python
# Inciailizamos los valores
maxima = temperaturas[0]
minima = temperaturas[0]

# Se buscan los valores en la lista de datos
for valor in temperaturas:

    # Obtener valor maximo
    if valor > maxima:
        maxima = valor
    
    # Obtener valor minimo
    if valor < minima:
        minima = valor

print("Temperatura Maxima:", maxima)
print("Tempertatura Minima:", minima)
```

La temperatura maxima es 15°C y la minima es 12°C, lo que deja ver poca variacion durante el año

## ¿Cuántos meses estuvieron por debajo de 14°C?

```python
# Se inicializa el contador para saber la cantidad de meses
contador = 0

for temp in temperaturas:

    # Se busca cuantas meses estan por debajo de 14 grados
    if temp < 14:

        # Se suma un numero por cada mes
        contador += 1

print("Meses por debajo de 14°C:", contador)
```

5 meses del año estuvieron por debajo de 14 grados, reflejando que hubo cierta variacion de clima
